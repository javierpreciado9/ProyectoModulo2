```python
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n=1000
def TragaMonedas():
    lista = ["diamante","campana","trebol","BAR"]

    a = random.choices(lista)
    b = random.choices(lista)
    c = random.choices(lista)
    d = random.choices(lista)
    e = random.choices(lista)
    f = random.choices(lista)

    return a,b,c,d,e,f
```

```python
for i in range(n):
    print (TragaMonedas(),'\n')
```

```python
 def juego():
    a,b,c,d,e,f= TragaMonedas()
    if (a==e and a==c):
        return True
    elif (d==b and d==f):
        return True
    elif (a==b and a==c and a==d and a==e and a==f):
        return True
    elif (d==b and d==e and d==c):
        return True
    elif (a==d and a==b and a==f and a==c):
        return True
    elif (d==e and d==b and d==c):
        return True
    elif(a==b and a==e and a==f):
        return True
    elif(d==a and d==e and d==c and d==f):
        return True
    else:
        return False
        
```

```python
cont_ganar = 0 
cont_perder = 0
for i in range (n):
    jugada = juego()
    if jugada:
        cont_ganar +=1
    else:
        cont_perder +=1
print("Se ganaron ",cont_ganar, " veces, y se perdieron ",cont_perder," veces.")
```

```python
def Tragamonedas(cap_inicial,apuesta,n_apuesta):
    capital = cap_inicial
    
    for i in range(n_apuesta):
        jugada = juego()
        if jugada:
            capital += apuesta
        else:
            capital -= apuesta
    return capital
```

```python
for i in range (1000):
    print(Tragamonedas(10000,200,150))
```
