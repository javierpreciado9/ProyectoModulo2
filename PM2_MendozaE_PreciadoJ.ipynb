## Modelo de combinaciones representado en el proyecto







```python
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n=10000
def TragaMonedas():
    lista = ["diamante","campana","trebol","BAR"]

    a = random.choices(lista)
    b = random.choices(lista)
    c = random.choices(lista)
    d = random.choices(lista)
    e = random.choices(lista)
    f = random.choices(lista)

    return a,b,c,d,e,f
```

```python
for i in range(n):
    print (TragaMonedas(),'\n')
```

```python
 def juego1():
    a,b,c,d,e,f= TragaMonedas()
    if (a==e and a==c):
        return True
    elif (d==b and d==f):
        return True
    elif (a==b and a==c and a==d and a==e and a==f):
        return True
    elif (d==b and d==e and d==c):
        return True
    elif (a==d and a==b and a==f and a==c):
        return True
    elif (d==e and d==b and d==c):
        return True
    elif(a==b and a==e and a==f):
        return True
    elif(d==a and d==e and d==c and d==f):
        return True
    else:
        return False
        
```

```python
cont_ganar = 0 
cont_perder = 0
for i in range (n):
    jugada = juego1()
    if jugada:
        cont_ganar +=1
    else:
        cont_perder +=1
print("Se ganaron ",cont_ganar, " veces, y se perdieron ",cont_perder," veces.")
```

```python
def apostador(cap_inicial,apuesta,n_apuesta):
    capital = [cap_inicial]
    
    for i in range(n_apuesta):
        jugada = juego()
        if jugada:
            capital.append(capital[-1] + apuesta)
        else:
            capital.append(capital[-1] - apuesta)
    return capital
```

```python
apostador(10000,200,150)
```

```python
def apostador2(cap_inicial,apuesta,n_apuesta):
    capital = cap_inicial
    
    for i in range(n_apuesta):
        jugada = juego()
        if jugada:
            capital += apuesta
        else:
            capital -= apuesta
    return capital
```

```python
for i in range (100):
    print(apostador2(10000,200,150))
```

```python
plt.figure(figsize=(10,8))

for i in range(100):
    capital=apostador(100,10,10)
    plt.plot(capital)

plt.xlabel('Numero de apuestas')
plt.ylabel('Capital')
plt.grid()
plt.show()
```
